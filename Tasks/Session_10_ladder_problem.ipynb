We wish to find the maximum length straight ladder than can fit around the provided corner.   
We must therefor find the parameters of the line ($y = mx + b$) that passes through the point $(-2,1)$ and maximizes $\text{dis} = \sqrt{x_i^2 + y_i^2}$. Where $x_i$ and $y_i$ are the respective slope intercept points.   
from the three points in question we have: 
\begin{align}
    (-1,2) &\rightarrow 2 = -m + b  & (x_i,0) &\rightarrow 0 = x_im + b  &  (0,y_i) &\rightarrow y_i = b  
\end{align}
Through some algebra we re-write the intercepts in terms of the slope
\begin{align}
    y_i &= 1 + 2m & x_i & = -\frac{1}{m} - 2   
\end{align}
Thus we have:
\begin{align}
    \text{dis} = \sqrt{\left( -\frac{1}{m} - 2\right)^2 + (1 + 2m)^2} 
    = \sqrt{4m^2 + 4m + 5 + \frac{4}{m} + \frac{1}{m^2}}
    = \sqrt{\frac{(2m+1)^2(m^2+1)}{m^2}}
    = \left|\frac{2m + 1}{m}\right| \sqrt{m^2 + 1}
\end{align}
We know obtain the first order derivative with respect to $m$ 
\begin{gather}
    \dfrac{\left(2m+1\right)\left(2m^3-1\right)}{m\sqrt{m^2+1}\left|m\right|\left|2m+1\right|}
\end{gather}
Which is $0$ at $m = \frac{1}{\sqrt[3]{2}}$. Thus we have that the maximum length straight ladder than can fit around the provided corner is 
\begin{gather}
    \text{dis}\left(\frac{1}{\sqrt[3]{2}}\right) \approx 4.162
\end{gather}

```python
import numpy as np
import matplotlib.pyplot as plt

def dis(m):
    """
    function takes in slope values and returns associated distance
    """
    return np.abs((2*m + 1)/m) * np.sqrt(m**2 + 1)

# Generate values for m
m = np.linspace(0.01, 100, 1_000)

# Calculate the function values
l = dis(m)

# Create the plot
plt.plot(m, l)
plt.title('Length of Ladder')
plt.xlabel('m')
plt.ylabel('dis(m)')
plt.grid(True)
plt.show()
```
