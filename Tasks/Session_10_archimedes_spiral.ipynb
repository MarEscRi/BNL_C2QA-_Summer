We know that the length of any given curve is given by: 
\begin{align}
    L &= \int_a^b \text{d}s & \text{d}s &= \sqrt{\text{d}x^2 + \text{d}y^2}
\end{align}
The archimedes spiral is defined in polar coordinates as $r = \theta$, as such it behooves us to know that: 
\begin{align}
    x &= r \cos{\theta} & y &= r \sin{\theta}
\end{align}
Additionally, we know that for any multivairable function the following is true: 
\begin{align}
    \text{d}f = \partial_u f \: \text{d}u + \partial_v f \: \text{d}v
\end{align}
Therefore we have: 
\begin{align}
    \text{d}x &= \cos{\theta} \: \text{d}r - r \sin{\theta} \: \text{d}\theta & 
    \text{d}y &= \sin{\theta} \: \text{d}r - r \cos{\theta} \: \text{d}\theta
\end{align}
Given that $r = \theta$, 
\begin{align}
    \text{d}x &= (\cos{\theta}  - r \sin{\theta}) \: \text{d}\theta & 
    \text{d}y &= (\sin{\theta}  - r \cos{\theta}) \: \text{d}\theta
\end{align}
Squaring and adding these terms we have that: 
\begin{gather}
    \text{d}s = \sqrt{1 + r^2} \: \text{d}r
\end{gather}
As such, the arc length is 
\begin{align}
    L &= \int_0^{8\pi} \sqrt{1 + r^2} \: \text{d}r 
\end{align}
Which can be solved by trigonometric substitution, but for our purposes we make use of [Integral-calculator](https://www.integral-calculator.com)
\begin{align}
    L &= \int_0^{8\pi} \sqrt{1 + r^2} \: \text{d}r = \dfrac{\operatorname{arsinh}\left(8{\pi}\right)+8{\pi}\sqrt{64{\pi}^2+1}}{2} \approx 318.03609
\end{align}


```python
"""archimedes_spiral.ipynb"""
# Cell 2

from scipy.integrate import quad  # type: ignore
import numpy as np 

def ds(r: float) -> float: 
    """
    displacement differential function that is meant to be integrated
    """
    return np.sqrt( 1 + r ** 2)

# Arc length
L: float = quad(ds, 0 , 8 * np.pi )[0]

print(rf"the archimedes spiral from 0 to 25.13 has arc length {L}")
print("")

theta = np.linspace(0, 8*np.pi, 1000)
radius = theta

fig = plt.figure()
ax = fig.add_subplot(111, polar=True)  

ax.plot(theta, radius, color='blue')  
ax.set_rticks([2 * np.pi, 4 * np.pi, 6 * np.pi, 8 * np.pi])  
ax.spines['polar']  


ax.set_yticks(np.arange(0, np.max(radius), 1), minor=True)
ax.yaxis.grid(True, alpha=0.5)

plt.title('Archimedes Spiral')
plt.show()
```
