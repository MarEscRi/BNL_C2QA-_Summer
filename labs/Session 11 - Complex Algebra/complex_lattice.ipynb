```python
"""complex_lattice.ipynb"""
# cell 1

import numpy as np
import matplotlib.pyplot as plt

# Define the complex function f(z)
def f(z):
    return z**2 + 1j*z + 1

# Generate input points within the specified range
x = np.linspace(-4, 4, 30)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)
Z = X + 1j*Y

# Apply the function f(z) to the input points
input_points = Z.flatten()
output_points = f(input_points)

# Extract the real and imaginary parts of the input and output points
input_real_part = np.real(input_points)
input_imaginary_part = np.imag(input_points)
output_real_part = np.real(output_points)
output_imaginary_part = np.imag(output_points)

# Create a scatter plot of the complex function
plt.scatter(input_real_part, input_imaginary_part, c='red', alpha=0.6, s=5, label='z')
plt.scatter(output_real_part, output_imaginary_part, c='blue', alpha=0.6, s=5, label='f(z)')
plt.title('Scatter Plot of f(z) = z^2 + iz + 1')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.xlim(-6, 18)
plt.ylim(-21, 21)
plt.legend()
plt.grid(True)
plt.show()

# Chatgpt was used for this implementation 
```

```python
"""complex_lattice.ipynb"""
# cell 2

import numpy as np

# Define the complex function f(z)
def f(z):
    return z**2 + 1j*z + 1

# Define the maximum absolute values for real and imaginary parts of f(z)
max_real_abs = 10
max_imag_abs = 10

# Initialize the count of satisfying Gaussian Integers
count = 0

# Iterate over all possible Gaussian Integers
for real_part in range(-max_real_abs, max_real_abs + 1):
    for imag_part in range(-max_imag_abs, max_imag_abs + 1):
        z = real_part + 1j*imag_part
        if np.imag(z) > 0 and np.abs(np.real(f(z))) <= max_real_abs and np.abs(np.imag(f(z))) <= max_imag_abs:
            count += 1

# Display the count of satisfying Gaussian Integers
print("The number of Gaussian Integers satisfying the conditions is:", count)

# Chatgpt was used for this implementation 
```
