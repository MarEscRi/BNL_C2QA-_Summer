```python
"""deutsch_qasm.ipynb"""

from __future__ import annotations

import typing

import numpy as np

from qiskit import Aer, QuantumCircuit, execute  # type: ignore
from qiskit.visualization import plot_bloch_multivector  # type: ignore
from qiskit.visualization import plot_distribution  # type: ignore
from qiskit.visualization import plot_state_qsphere  # type: ignore
from qis101_utils import as_latex

if typing.TYPE_CHECKING:
    from typing import Any


# from qiskit.visualization import plot_histogram

backend: Any = Aer.get_backend("qasm_simulator")


# define the Deutsch oracle circuit
def deutsch_oracle(qc:Any) -> None:
    qc.i(0)
    qc.i(1)
    #qc.cx(0, 1)
    #qc.z(0)
    


# create a two-qubit quantum circuit
qc: Any = QuantumCircuit(2, 1)

qc.x(1)

# apply a Hadamard gate to the input qubit
qc.h(0)

# apply the Deutsch oracle
deutsch_oracle(qc)

# apply another Hadamard gate to the input qubit
qc.h(0)

# measure the input qubit and output the result
qc.measure(0, 0)

# simulate the circuit using the statevector simulator

result: Any = execute(qc, backend).result()
counts: Any = result.get_counts(qc)

display(qc.draw(output="mpl", scale=1.5))

# plot the results on a histogram
plot_distribution(counts)
```
